# **Fruit Image Classifier**

This file contains the complete pipeline developed by our team, covering the entire machine learning lifecycle: from initial data diagnostic analysis to baseline construction, iterative model optimization, and comparative transfer learning studies.

---

## **Project Structure & Module Overview**

### 01_dataset_analysis
* **Purpose:** Initial data auditing and Exploratory Data Analysis .
* **Key Actions:** Verified environment dependencies and dataset integrity; analyzed class distribution and image spatial dimensions; identified and flagged mislabeled samples to ensure data quality before training.

### 02_baseline_model
* **Purpose:** Establishing a performance benchmark.
* **Configuration:** A lightweight CNN featuring 2 convolutional layers, 64x64 input resolution, and 10 training epochs.
* **Goal:** To validate the end-to-end training/testing pipeline.

### 03_architecture_depth_study
* **Purpose:** Evaluating the impact of network depth on feature extraction.
* **Experiment:** Extended training to 30 epochs and compared performance gains when scaling from 2 to 4 convolutional layers.

### 04_input_resolution_study
* **Purpose:** Investigating the relationship between spatial information and accuracy.
* **Experiment:** Tested a range of input resolutions (from 16x16 up to 128x128) to determine the optimal balance between computational cost and model precision.

### 05_regularization_study
* **Purpose:** Enhancing model robustness and combating overfitting.
* **Techniques:** Integrated Dropout layers and Batch Normalization (BatchNorm) to improve convergence speed and generalization performance on unseen data.

### 06_class_balance & data_augmentation
* **Purpose:** Addressing data distribution skewness and enhancing diversity.
* **Strategies:** Implemented Class-Weighted Loss and Early Stopping; utilized data augmentation and synthetic image blending to balance the 4 target classes. Applied CosineAnnealingLR (learning rate scheduling) for smoother convergence.

### 07_model_optimization
* **Purpose:** Final refinement of the custom architecture.
* **Outcome:** Presents two optimized versions of our model, integrating the best-performing hyperparameters and structural configurations identified in previous notebooks.

### 08_transfer_learning
* **Purpose:** Benchmarking against State-of-the-Art architectures.
* **Methodology:** A rigorous comparative study of ResNet50, EfficientNet-B3, and ConvNeXt-Small.
* **Advanced Features:** Includes strict training isolation for reproducible results, Test-Time Augmentation (TTA), and sophisticated post-processing logic to maximize classification accuracy.
